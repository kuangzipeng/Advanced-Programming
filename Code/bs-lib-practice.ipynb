```python
import requests
def gethtml(url):
    try:
        r=requests.get(url)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return "异常"
    
    
url="http://yjsy.blcu.edu.cn/col/col13205/index.html"
print(gethtml(url))
    
```
