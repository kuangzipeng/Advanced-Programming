```python
class Leaf(object):
    def __init__(self, color="green"):
        self.color = color
    def fall(self):
        print ("Splat!")
```

```python
leaf = Leaf()
```

```python
class MapleLeaf(Leaf):
    def change_color(self):
        if self.color == "green":
            self.color = "red"
```

```python
mleaf= MapleLeaf()
```

```python
class MapleLeaf(Leaf):
    def change_color(self):
        if self.color == "green":
            self.color = "red"
    def fall(self):
        self.change_color()
        print ("Plunk!")
```

```python
mleaf1= MapleLeaf()
```
